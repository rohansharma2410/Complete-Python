```python
import cv2
import pytesseract
from gtts import gTTS

# Load the image
img = cv2.imread('Screenshot (21).png')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Perform text recognition using Tesseract OCR
pytesseract.pytesseract.tesseract_cmd = r'C:/Users/Rohan/Desktop/Python/env/Tesseract-OCR' # Update this path to the location of your Tesseract OCR executable
text = pytesseract.image_to_string(gray)

# Speak the recognized text using gTTS
speech = gTTS(text=text, lang='en')
speech.save("text.mp3")

# Play the audio file
import os
os.system("mpg321 text.mp3")
```
